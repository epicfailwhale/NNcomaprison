Autoencoder from https://www.python-engineer.com/posts/pytorch-autoencoder/

```python
import numpy
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
import matplotlib 
import matplotlib.pyplot as plt
```

```python
transform = transforms.ToTensor()
# converts data range 0-255 in numpyarray, to 0-1 in tensor 'array'
# a pytorch tensor - basically the same thing as a numpy array


# transform = transforms.Compose([
#     transforms.ToTensor(),
#     transforms.Normalize((0.5),(0.5))
# ])

mnist_data = datasets.MNIST(root = './data', train=True, download=True, transform=transform)
# MNIST data already included in pytorch
# root -> create a folder named data 
# download and stores the data in there
# data is not in csv form but in ubyte form:
# 8bit unsigned int data, values ranging from 0 to 255

data_loader = torch.utils.data.DataLoader(dataset=mnist_data, batch_size=64, shuffle=True)
```

```python
dataiter = iter(data_loader) 
# dataiter will iterate through the elements within data_loader
images, labels = dataiter.next()
# pulls out images and their labels -> .next moves to next element

print(torch.min(images), torch.max(images))
# see that all the values of each pixel is between 0 and 1
# because transforms in mnist_data is totensor:
# this can be changed, so it won't always be b/w 0 and 1
# eg. using transforms.compose -> gets [-1,0]
```

```python
# creating autoencoder class

class Autoencoder(nn.Module):
    def __init__(self):
        
        super().__init__()
        # lets you access methods from parent class within child class
        # total of all images have size N (batch size), 784 (28x28)
        
        # now reduce with linear layers
        
        # sequential model - runs all of the modules repeatedly
        self.encoder = nn.Sequential(
            nn.Linear(28*28, 128), # N,784 -> reduce to N,128
            nn.ReLU(), # activation function
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 12),
            nn.ReLU(),
            nn.Linear(12, 3) # final output size is N, 3
        ) #ie. 3 pixels
        
        
        # decoder - the opposite direction, create larger output
        self.decoder = nn.Sequential(
            nn.Linear(3, 12), # N,3 -> increase to N,784
            nn.ReLU(), # activation function
            nn.Linear(12, 64),
            nn.ReLU(),
            nn.Linear(64, 128),
            nn.ReLU(),
            nn.Linear(128, 28*28), # then needs activation function to put
            nn.Sigmoid()                 # values back into the range 0-1
        ) 
        
        pass
    
        
    # the forward pass    
    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded
    

    # notes: input images range [-1,1] -> nn.Tanh as final activation function
```

```python
# yes this is just setting up training
model = Autoencoder()
criterion = nn.MSELoss() # mean squared error loss
optimizer = torch.optim.Adam(model.parameters(), lr = 1e-3, weight_decay = 1e-5) 
# optimises the model using algorithms
```

```python
# training dataset
num_epochs = 10
outputs = []
for epoch in range(num_epochs):
    for (img,_) in data_loader:
        img = img.reshape(-1, 28*28)
        recon = model(img)
        loss = criterion(recon, img)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    print(f'Epoch:{epoch+1}, Loss:{loss.item():.4f}') # f makes it easier to type out stuff with variables
    # instead of always separating them all with ''
    
    outputs.append((epoch, img, recon))
#   print('Epoch:{}, Loss:{:.4f}'.format(epoch+1, float(loss)))
#   outputs.append((epoch, img, recon),)
```

```python
# this is the performance of the NN at various points in training
# imgs is a (32,784) array, at each [k], you get all of the original images
# recon is also (32, 784), get the reconstructed images after compression

for k in range(0, num_epochs):
    imgs = outputs[k][1].detach().numpy() # get the original images
    recon = outputs[k][2].detach().numpy()
   
    
for i in range(0,len(imgs)):
    one_image = numpy.asfarray(imgs[i].reshape((28,28)))    # the list is made into an array
    matplotlib.pyplot.imshow(one_image, cmap = matplotlib.pyplot.get_cmap('Reds')) # some options
    matplotlib.pyplot.show()
    
    recon_image = numpy.asfarray(recon[i].reshape((28,28)))
    matplotlib.pyplot.imshow(recon_image, cmap = matplotlib.pyplot.get_cmap('Blues'))
    matplotlib.pyplot.show()
    

    
```

```python
# what does the forward pass do?!
num_epochs = 10
outputs = []
for epoch in range(num_epochs):
    for (img,_) in data_loader:
        img = img.reshape(-1, 28*28)
        recon = model.forward(img)
        loss = criterion(recon, img)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
    print(f'Epoch:{epoch+1}, Loss:{loss.item():.4f}') # f makes it easier to type out stuff with variables  
    
    outputs.append((epoch, img, recon))
```
